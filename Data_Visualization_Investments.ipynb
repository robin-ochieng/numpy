#South Africa


```python
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter

# Data
year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
inflation_rate = [4.09, 5.00, 5.72, 5.78, 6.13, 4.54, 6.57, 5.18, 4.52, 4.12, 3.21, 4.61, 6.90]

# Define a function to add the percent sign
def percent_format(x, pos):
    return '{:.1f}%'.format(x)

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(13, 7))

# Plot the data and add labels for each point
for i in range(len(year)):
    ax.plot(year[i], inflation_rate[i], 'o', color='blue', markersize=8)
    ax.text(year[i], inflation_rate[i] + 0.2, str(inflation_rate[i])+'%', horizontalalignment='center', fontsize=12)

# Connect the dots with a line
ax.plot(year, inflation_rate, color='blue', linewidth=2)

# Add a title and labels to the axis
ax.set_title('South Africa Inflation Rate', fontsize=18)
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Inflation rate', fontsize=14)

# Add the percent sign to the y-axis labels
formatter = FuncFormatter(percent_format)
ax.yaxis.set_major_formatter(formatter)

# Show the plot
plt.show()

```

```python
import numpy as np
from scipy.interpolate import make_interp_spline

# Data
year = [2010, 2011, 2012, 2013, 2014, 2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
inflation_rate = [4.09, 5.00, 5.72, 5.78, 6.13, 4.54, 6.57, 5.18, 4.52, 4.12, 3.21, 4.61, 6.90]

# Create a smoother curve using spline interpolation
x_new = np.linspace(min(year), max(year), 300)
spl = make_interp_spline(year, inflation_rate, k=3)  # k=3 specifies a cubic spline interpolation
y_new = spl(x_new)

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(13, 7))

# Plot the data points and the smoothed curve
ax.plot(year, inflation_rate, 'o', color='blue', markersize=8)
ax.plot(x_new, y_new, color='blue', linewidth=2)

# Add a title and labels to the axis
ax.set_title('South Africa Inflation Rate', fontsize=18)
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Inflation rate', fontsize=14)

# Add the percent sign to the y-axis labels
formatter = FuncFormatter(percent_format)
ax.yaxis.set_major_formatter(formatter)

# Show the plot
plt.show()
```

```python
import matplotlib.pyplot as plt

# Data
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
bond_rate = [10.60, -3.39, 14.57, 10.18, 7.36, 10.60, 7.90, 8.54]

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(13, 7))

# Plot the data
ax.plot(year, bond_rate, color='red', linewidth=2)

# Add values for each year
for i in range(len(year)):
    ax.text(year[i], bond_rate[i], str(bond_rate[i]) + '%', ha='center', va='bottom', fontsize=12)

# Add a title and labels to the axis
ax.set_title('South Africa Bond Rate', fontsize=18)
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Bond Rate (%)', fontsize=14)

# Show the plot
plt.show()
```

```python
import numpy as np
from scipy.interpolate import make_interp_spline

# Data
year = [2015, 2016, 2017, 2018, 2019, 2020, 2021, 2022]
bond_rate = [10.60, -3.39, 14.57, 10.18, 7.36, 10.60, 7.90, 8.54]

# Interpolate the data
xnew = np.linspace(min(year), max(year), 300)
spl = make_interp_spline(year, bond_rate, k=3)
bond_rate_smooth = spl(xnew)

# Create a figure and axis object
fig, ax = plt.subplots(figsize=(13, 7))

# Plot the smoothed data
ax.plot(xnew, bond_rate_smooth, color='red', linewidth=2)

# Add values for each year
for i in range(len(year)):
    ax.text(year[i], bond_rate[i], str(bond_rate[i]) + '%', ha='center', va='bottom', fontsize=12)

# Add a title and labels to the axis
ax.set_title('South Africa Bond Rate', fontsize=18)
ax.set_xlabel('Year', fontsize=14)
ax.set_ylabel('Bond Rate (%)', fontsize=14)

# Show the plot
plt.show()
```
